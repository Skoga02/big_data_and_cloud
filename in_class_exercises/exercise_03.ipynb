```python
# Register a dataframe for use in SQL
df_nobel_prize.createOrReplaceTempView("df_nobel_prize")

# TODO: pick out swedish medals in table tennis
spark.sql("""
          SELECT
            DISTINCT category
          FROM df_nobel_prize
          
          """).display()
```

```python
df_swe_award_winners = spark.sql("""
          SELECT
            name,
            birth_country,
            COUNT(*) OVER (PARTITION BY birth_country) AS country_medal_count
          FROM df_nobel_prize
          WHERE birth_country = 'Sweden'
          """).display()
```

# Short EDA with a graph

```python
df_swe_winners = spark.sql("""
          SELECT
            name, count(*) AS award_count
          FROM df_nobel_prize
          WHERE birth_country = 'Sweden'
          AND category IS NOT NULL
          GROUP BY name
          ORDER BY award_count DESC
          """)
df_swe_winners.display()
```

```python
df_swe_winners.plot(kind = "bar", y="name", x = "award_count")
```
